## a

$E(X) = \mu$

$Var(X) = \frac{\sigma^2}{n}$


${X_1}, \dots, {X_{16}} \sim N(262.5, 12^2)$

$\overline{x} \sim N(262.5, (\sqrt{\frac{12^2}{16}})^2)$

${Y_1}, \dots, {Y_{16}} \sim N(262.5, 12^2)$

$\overline{y} \sim N(262.5, (\sqrt{\frac{12^2}{16}})^2)$

$d = (\overline{x} - \overline{y}) \sim N(0, (\sqrt{\frac{12^2}{16} + \frac{12^2}{16}})^2)$

$P(|\overline{x} - \overline{y}|<5) = P(-5 < d < 5)$ 

```python
mean = 0

std = sqrt((12**2/16) + (12**2/16))

norm.cdf(x=5, loc=mean, scale=std) - norm.cdf(x=-5, loc=mean, scale=std)
```
